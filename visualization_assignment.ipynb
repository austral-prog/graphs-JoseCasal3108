<h1><b> Temporadas y ratings de "Better Call Saul"</b></h1>

```python
# Datos: Temporadas y su calificación promedio aproximada en IMDb
temporadas = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6']
calificaciones = [8.4, 8.7, 8.8, 8.9, 9.2, 9.6]

plt.figure(figsize=(9, 5))

# Trazo la línea con marcadores
plt.plot(temporadas, calificaciones, marker='o', markersize=8, color='yellow', 
         linewidth=3, markeredgecolor='black', label='Calificación Promedio')

# Personalización
plt.title('Evolución de las Calificaciones de Better Call Saul (IMDb)', fontsize=14, fontweight='bold')
plt.xlabel('Temporada')
plt.ylabel('Calificación (0 - 10)')
plt.ylim(8.0, 10.0) # Ajustar el eje Y para resaltar los cambios

# Añadir los valores exactos sobre cada punto
for i, calif in enumerate(calificaciones):
    plt.text(i, calif + 0.05, str(calif), ha='center', fontweight='bold')

plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.legend(loc='upper left')

plt.savefig('imgs/bcs_ratings_line.png', dpi=300, bbox_inches='tight')
plt.show()
```

<figcaption><i> En este grafico podemos ver como las calificaciones de la serie Better Call Saul en netflix (datos sacados de la pagina IMDb) van cambiando segun las temporadas, las ultimas teniendo bastante mejor rating q las primeras</i></figcaption>

<h1><b> Velocidades máximas según el golpe de Ping Pong</b></h1>

```python
# Datos (velocidades aproximadas en km/h)
golpes = ['Smash (Remate)', 'Topspin Fuerte', 'Drive Rápido', 'Loop (Efecto)', 'Push/Cortada', 'Bloqueo']
velocidades = [110, 85, 70, 50, 25, 15]

plt.figure(figsize=(10, 6))

# Creo las barras horizontales
barras = plt.barh(golpes, velocidades, color='blue', edgecolor='black')

# Añado los valores al final de cada barra
for barra in barras:
    ancho = barra.get_width()
    plt.text(ancho + 2, barra.get_y() + barra.get_height()/2, f'{ancho} km/h', va='center', fontweight='bold')

plt.title('Velocidad Máxima Típica por Tipo de Golpe en Ping Pong', fontsize=14)
plt.xlabel('Velocidad (km/h)')
plt.ylabel('Tipo de Golpe')

# Invierto el eje Y para que el golpe más rápido quede arriba
plt.gca().invert_yaxis()

plt.savefig('imgs/ping_pong_speeds.png', dpi=300, bbox_inches='tight')
plt.show()
```

<figcaption><i> En este grafico de tipo "Bar Chart" podemos observar las diferentes velocidades de cada golpe que se emplea en el ping pong</i></figcaption>

<h1><b> Relación de altura y alcance de remate en voley profesional</h1></b>

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulación de datos (basados en promedios del vóley profesional masculino)
np.random.seed(10)

# Generamos datos de (Estatura en cm, Alcance de remate en cm)
# Centrales (Middle Blockers): Los más altos, saltan mucho por que suelen ser mas altos
estatura_centrales = np.random.normal(205, 4, 30)
alcance_centrales = estatura_centrales + np.random.normal(140, 5, 30)

# Opuestos (Opposite Hitters): Altos, el ataque mas fuerte del equipo
estatura_opuestos = np.random.normal(200, 5, 25)
alcance_opuestos = estatura_opuestos + np.random.normal(145, 6, 25)

# Puntas (Outside Hitters): Buena tecnica y ataque constante
estatura_puntas = np.random.normal(195, 4, 35)
alcance_puntas = estatura_puntas + np.random.normal(135, 5, 35)

# Armadores (Setters): Suelen ser los mas bajos del equipo, no dependen tanto de la altura
estatura_armadores = np.random.normal(188, 5, 20)
alcance_armadores = estatura_armadores + np.random.normal(125, 4, 20)

# Configuración de la figura
plt.figure(figsize=(10, 7), facecolor='#ffffff')

# Trazar los puntos (Scatter)
# Uso un poco de transparencia (alpha) para ver puntos superpuestos
plt.scatter(estatura_centrales, alcance_centrales, c='#E63946', label='Central', s=80, alpha=0.7, edgecolors='black')
plt.scatter(estatura_opuestos, alcance_opuestos, c='yellow', label='Opuesto', s=80, alpha=0.7, edgecolors='black')
plt.scatter(estatura_puntas, alcance_puntas, c='green', label='Punta', s=80, alpha=0.7, edgecolors='black')
plt.scatter(estatura_armadores, alcance_armadores, c='blue', label='Armador', s=80, alpha=0.7, edgecolors='black')

# Títulos y etiquetas
plt.title('Relación entre Estatura y Alcance de Remate en Vóley', fontsize=15, fontweight='bold', pad=15)
plt.xlabel('Estatura del Jugador (cm)', fontsize=12)
plt.ylabel('Alcance Máximo de Remate (cm)', fontsize=12)

# Cuadrícula y leyenda
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(title='Posición en Cancha', title_fontsize='11', fontsize='10', loc='upper left')

# Exporto y muestro
plt.savefig('imgs/voley_scatter.png', dpi=300, bbox_inches='tight')
plt.show()
```

<figcaption><i> En este gráfico de tipo scatter plot vemos la relación de altura/alcance de remate de jugadores profesionales de voley según su posición</i></figcaption>
